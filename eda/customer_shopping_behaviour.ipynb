## Fill up missing values of Review Rating with median for each category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df['Review Rating'].isnull().sum()
```

```python
df[['Category','Review Rating']].head(20)
```

```python
df.isnull().sum()
```

# Convert all columns(camel case) into snake case

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

# Feature engineering

```python
# create a column age_goup
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age' , 'age_group']].head(20)
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
# create a column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(20)
```

### Check whether discount_applied and promo_code_used are equal if they are equal then drop one of among them

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis = 1)
```
